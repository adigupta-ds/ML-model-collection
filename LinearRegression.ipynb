```python
for col in df.columns:
  if df[col].dtype == "object":
    df[col] = df[col].fillna(df[col].mode([0]))
  else:
    df[col]=df[col].fillna(df[col].mean())
```

```python
import matplotlib.pyplot as plt

import seaborn as sns
```

```python
for col in df.columns:
  if(df[col].dtype != 'object'):
    sns.boxplot(df[col])
    plt.xlabel(col)
    plt.show()
```

```python
# Outlier Removal

outlist=['bmi','past_consultations','Hospital_expenditure','Anual_Salary']

for col in outlist:
  Q1 = df[col].quantile(0.25)
  Q3 = df[col].quantile(0.75)

  IQR = Q3 - Q1 # (Inter Quantile Range)

  LB = Q1 - 1.5*(IQR)
  UB = Q3 + 1.5*(IQR)

  df=df[(df[col]>= LB) & (df[col]<=UB)]
```

```python
x = df.iloc[:,:-1]
x

# x=df.drop(columns=['charges'])
```

```python
y=df['charges']
y
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x,y,train_size=0.80,random_state=0)

```

```python
# 100 Questions / Solutions

# i give my student 80 questions(x_train) with solutions(y-train)

#  i have 20 questions with correct solutions

# i am going to take a test

# i will give the students the set of 20 questions(x_test)

# the students will give some answer(correct/incorrect)(y_pred)

# i will compare the students's answer with the correct answers that i have with me

# compare y_test with y_pred
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
y_pred
```

```python
result = pd.DataFrame(columns = ['Actual','Predicted'])
result['Actual'] = y_test
result['Predicted'] = y_pred

result
```

```python
from sklearn.metrics import *

res = r2_score(y_test,y_pred)
res*100

# r2_score = how well your model is predicting the target
```
